Датасет где нужно узнать доход человека на основе данных о нём, предоставленных ниже. Он выше или меньше 50K $

```python
import pandas as pd

df = pd.read_csv('adult.csv')

df.head(5)
```

```python
categorial_features = ['State-gov', 'Bachelors', 'Never-married', 'Adm-clerical', 'Not-in-family', 'White', 'Male', 'United-States']
```

```python
normal_values = []

for value in df[' <=50K']:
    if value == " <=50K":
        normal_values.append(0)
    if value == " >50K":
        normal_values.append(1)

df[' <=50K'] = normal_values
```

```python
df[' <=50K'].value_counts()
```

```python
df = pd.get_dummies(df, prefix=categorial_features)
```

```python
df.describe()
```

```python
df.head(5)
```

4. обучить любой классификатор (какой вам нравится)

```python
X = df.drop(columns=[' <=50K'])
y = df[' <=50K']
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

params_model = {
    'eval_metric': 'F1',
    'auto_class_weights': 'Balanced',
    'silent': True,
    'one_hot_max_size': 20,
    'early_stopping_rounds': 50,
    'boosting_type': 'Ordered',
    'allow_writing_files': False
}

cbr_final_model = CatBoostClassifier(
    **params_model,
    depth=8,
    iterations=100,
    learning_rate=0.5
)

cbr_final_model.fit(X_train, y_train, eval_set=(X_test, y_test))
y_predict = cbr_final_model.predict(X_test)
```

```python
from sklearn.metrics import recall_score, precision_score, roc_auc_score, accuracy_score, f1_score

def evaluate_results(y_test, y_predict):
    print('Classification results:')
    f1 = f1_score(y_test, y_predict)
    print("f1: %.2f%%" % (f1 * 100.0)) 
    roc = roc_auc_score(y_test, y_predict)
    print("roc: %.2f%%" % (roc * 100.0)) 
    rec = recall_score(y_test, y_predict, average='binary')
    print("recall: %.2f%%" % (rec * 100.0)) 
    prc = precision_score(y_test, y_predict, average='binary')
    print("precision: %.2f%%" % (prc * 100.0)) 

    
evaluate_results(y_test, y_predict)
```

6. применить random negative sampling для построения классификатора в новых условиях

```python
mod_data = df.sample(frac=1)

neg_sample = mod_data[mod_data[' <=50K']==0][:len(mod_data[mod_data[' <=50K']==1])]
sample_test = mod_data[mod_data[' <=50K']==0][len(mod_data[mod_data[' <=50K']==1]):]

pos_sample = mod_data[mod_data[' <=50K']==1]
print(neg_sample.shape, pos_sample.shape)

sample_train = pd.concat([neg_sample, pos_sample]).sample(frac=1)
```

```python
cbr_final_model.fit(sample_train.iloc[:,:-2].values, 
          sample_train.iloc[:,-2].values)
y_predict = cbr_final_model.predict(sample_test.iloc[:,:-2].values)
evaluate_results(sample_test.iloc[:,-2].values, y_predict)
```
